```python
import numpy as np

months = np.array([
    "Jan","Feb","Mar","Apr","May","Jun",
    "Jul","Aug","Sep","Oct","Nov","Dec"
])

# Sample data (in $1000s)
sales = np.array([45, 50, 48, 55, 60, 58, 62, 65, 59, 63, 70, 75])

total_sales = sales.sum()
avg_sales = sales.mean()
highest_sales = sales.max()
lowest_sales = sales.min()
best_month = months[sales.argmax()]
worst_month = months[sales.argmin()]
above_avg_sales = (sales > avg_sales).sum()
below_avg_sales = (sales < avg_sales).sum()

print("Total Sales:", total_sales * 1000)
print("Average Sales:", avg_sales * 1000)
print("Highest Sales:", highest_sales * 1000)
print("Lowest Sales:", lowest_sales * 1000)
print("Best Month:", best_month)
print("Worst Month:", worst_month)
print("Number of Months Above Average:", above_avg_sales)
print("Number of Months Below Average:", below_avg_sales)
```

```python
bonus = 2  # in $1000s
sales_with_bonus = sales + bonus
print("Sales after bonus:", sales_with_bonus * 1000)
```

```python
cost = 5
net_sales = sales - cost
print("Net sales after cost:", net_sales * 1000)
```

```python
growth_factor = 1.10
grown_sales = sales * growth_factor
print("Sales after 10% growth:", grown_sales * 1000)
```

```python
total_sales = sales.sum()
sales_share = sales / total_sales
print("Monthly sales share:", sales_share)
```

```python
final_sales = (sales + 2) * 1.10 - 5
print("Final adjusted sales:", final_sales * 1000)
```
